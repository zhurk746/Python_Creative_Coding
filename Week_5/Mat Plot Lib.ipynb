```python
# Creating 3 random charts to plot the amount of clothes both my boyfriend and I own to prompt a conversation about how to best split closet space.
```

```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
s = "Alex Clothing Count|Kasey Clothing Count".split("|")

# Generate dummy data into a dataframe
j = {x: [random.choice(["Jeans", "T-Shirts", "Long-Sleeves", "Sweaters", "Socks", "Sweatpants", "Sweatshirts", "Pajamas"]
                       ) for j in range(50)] for x in s}
df = pd.DataFrame(j)

index = np.arange(8)
bar_width = 0.35

fig, ax = plt.subplots()
alex = ax.bar(index, df["Alex Clothing Count"].value_counts(), bar_width,
                label="Alex's Clothes", color='blue')

kasey = ax.bar(index+bar_width, df["Kasey Clothing Count"].value_counts(),
                 bar_width, label="Kasey's Clothes", color='green')

ax.set_xlabel('Clothing Type')
ax.set_ylabel('Count')
ax.set_title("Kasey vs Alex Closet Battle")
ax.set_xticks(index + bar_width/2)
ax.set_xticklabels(["Jeans", "T-Shirts", "Long-Sleeves", "Sweaters", "Socks", "Sweatpants", "Sweatshirts", "Pajamas"], rotation=45)
ax.legend()

plt.show()
```

```python
s = "Alex Clothing Count|Kasey Clothing Count".split("|")

# Generate dummy data into a dataframe
j = {x: [random.choice(["Jeans", "T-Shirts", "Long-Sleeves", "Sweaters", "Socks", "Sweatpants", "Sweatshirts", "Pajamas"]
                       ) for j in range(50)] for x in s}
df = pd.DataFrame(j)

index = np.arange(8)

clothing=("Jeans", "T-Shirts", "Long-Sleeves", "Sweaters", "Socks", "Sweatpants", "Sweatshirts", "Pajamas")
y_pos = np.arange(len(clothing))

# plot lines
plt.plot(index, df["Alex Clothing Count"].value_counts(), label = "Alex Clothes", color='blue')
plt.plot(index, df["Kasey Clothing Count"].value_counts(), label = "Kasey Clothes", color='green')
plt.xlabel('Clothing Type')
plt.ylabel('Count')
plt.title("Kasey vs Alex Closet Battle")
plt.xticks(y_pos, clothing, rotation=45)
plt.legend()
plt.show()
```

```python
s = "Alex Clothing Count|Kasey Clothing Count".split("|")

# Generate dummy data into a dataframe
j = {x: [random.choice(["Jeans", "T-Shirts", "Long-Sleeves", "Sweaters", "Socks", "Sweatpants", "Sweatshirts", "Pajamas"]
                       ) for j in range(50)] for x in s}
df = pd.DataFrame(j)

index = np.arange(8)

clothing=("Jeans", "T-Shirts", "Long-Sleeves", "Sweaters", "Socks", "Sweatpants", "Sweatshirts", "Pajamas")
y_pos = np.arange(len(clothing))


#find line of best fit
a, b = np.polyfit(y_pos, df["Alex Clothing Count"].value_counts(), 1)
c, d=np.polyfit(y_pos, df["Kasey Clothing Count"].value_counts(), 1)

plt.scatter(y_pos, df["Alex Clothing Count"].value_counts(), label='Alex Clothes', color='blue')
plt.scatter(y_pos, df["Kasey Clothing Count"].value_counts(), label = "Kasey Clothes", color='green')
plt.plot(y_pos, a*y_pos+b, color='blue')
plt.plot(y_pos, c*y_pos+d, color='green')
plt.xlabel('Clothing Type')
plt.ylabel('Clothing Count')
plt.title("Kasey vs Alex Closet Battle")
plt.xticks(y_pos, clothing, rotation=45)
plt.legend()
plt.show()
```
