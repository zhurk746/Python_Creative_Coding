```python
import cv2
import numpy as np
import matplotlib.image as img
from matplotlib import pyplot as plt
```

```python
#reading the image
img = cv2.imread("images/Pioneer_Woman.jpg")
type(img)
```

```python
#Create Edge Mask
def edge_mask(img, line_size, blur_value):
  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  gray_blur = cv2.medianBlur(gray, blur_value)
  edges = cv2.adaptiveThreshold(gray_blur, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, line_size, blur_value)
  return edges
```

```python
line_size = 7
blur_value = 7
```

```python
edges = edge_mask(img, line_size, blur_value)
```

```python
filename = 'Pioneer_Edges.jpg'
# Using cv2.imwrite() method
# Saving the image
cv2.imwrite(filename, edges)
```

```python
import textwrap
```

```python
new_img = cv2.imread('Pioneer_Edges.jpg')
print(new_img.shape)

height, width, channel = new_img.shape

text_img = np.ones((height, width))
print(text_img.shape)
font = cv2.FONT_HERSHEY_TRIPLEX

text = "This is the best cookbook ever! Drummond | OpenAI"
wrapped_text = textwrap.wrap(text, width=35)
x, y = 10, 40
font_size = 1
font_thickness = 1

for i, line in enumerate(wrapped_text):
    textsize = cv2.getTextSize(line, font, font_size, font_thickness)[0]

    gap = textsize[1] + 10

    y = int((new_img.shape[0] + textsize[1]) / 2) + i * gap
    x = int((new_img.shape[1] - textsize[0]) / 2)

    cv2.putText(new_img, line, (x, y), font,
                font_size, 
                (148,252,3), 
                font_thickness, 
                lineType = cv2.LINE_AA)

cv2.imshow("Result Image", new_img)
```
